```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imshow

from glob import glob
import os
```

```python
MAIN_PATH = './data/finding-lungs-in-ct-data/'
FILE_PATH = os.path.join(MAIN_PATH, '3d_images')
```

```python
FILE_LIST = sorted(glob(os.path.join(FILE_PATH, '*')))
print("[전체 파일 리스트]")
FILE_LIST
```

```python
IMG_LIST = sorted(glob(os.path.join(FILE_PATH, 'IMG_*')))
MASK_LIST = sorted(glob(os.path.join(FILE_PATH, 'MASK_*')))
```

```python
print("[이미지 파일 리트스]")
IMG_LIST
```

```python
print("[마스트 파일 리스트]")
MASK_LIST
```

```python
import nibabel as nib

fig, ax = plt.subplots(1, 4, figsize=(10, 10))

for i in range(len(IMG_LIST)):
    test_image = nib.load(IMG_LIST[i]).get_data() #get_data덕분에 numpy_array로 변환됨
    print(test_image.shape)
    ax[i].imshow(test_image[test_image.shape[0]//2], cmap='bone')
    
plt.show()
    
```

```python
fig, ax = plt.subplots(1, 4, figsize=(10, 10))

for i in range(len(MASK_LIST)):
    test_mask = nib.load(MASK_LIST[i]).get_data()
    print(test_mask.shape)
    ax[i].imshow(test_mask[test_mask.shape[0]//2], cmap='gray')
plt.show()
```

```python
from skimage.util import montage

fig, ax = plt.subplots(1, 1, figsize=(20, 20))
ax.imshow(montage(test_image), cmap='bone')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 20))
ax.imshow(montage(test_mask), cmap='bone')
```
